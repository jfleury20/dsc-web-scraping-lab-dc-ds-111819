## Retrieving Titles

To start, write a function that extracts the titles of the books on a given page. The input for the function should be the `soup` for the HTML of the page.

````python
from bs4 import BeautifulSoup
import requests
````

````python
page = requests.get('http://books.toscrape.com/')
soup = BeautifulSoup(page.content, 'html.parser')
````

````python
print(soup.prettify())
````

````python
soup.select("div h3 a")[0]
# find all a tags within an h3 tag within a div tag. Then display first element.
print(type(soup.select("div h3 a")[0]['title']))
# display the 'title' class 
````

````python
# using .find instead of .select

soup.find_all('h3')[0]
# iterable object of all h3 tags

print(soup.find_all('h3')[0].find('a').attrs['title'])
# get the 'a' tag within the 'h3' tag using the .attrs attribute

print(soup.find_all('h3')[0].find('a')['title'])
# same thing without the .attrs attribute
````

````python
def retrieve_titles(soup):
    titles = [a['title'] for a in soup.select("div h3 a")]
    return titles
````

````python
retrieve_titles(soup)
````

## Retrieve Ratings

Next, write a similar function to retrieve the star ratings on a given page. Again, the function should take in the `soup` from the given HTML page and return a list of the star ratings for the books. These star ratings should be formatted as integers.

````python
print(soup.prettify())
````

````python
# find all 'star-rating' classes within a p tag
soup.find_all('p', class_='star-rating')
````

````python
print(soup.select("div p.star-rating"))
# find all 'star-rating' classes within a p tag within a div tag

soup.select("div p.star-rating")[0]['class']
# get only the class from the results
````

````python
def retrieve_ratings(soup):
    ratings = [r['class'][-1] for r in soup.select("div p.star-rating")]
    # get the last element in the ['star-rating', 'Three'] list
    
    return ratings
````

````python
retrieve_ratings(soup)
````

## Retrieve Prices

Now write a function to retrieve the prices on a given page. The function should take in the `soup` from the given page and return a list of prices formatted as floats.

````python
print(soup.prettify())
````

````python
soup.find_all('p', class_='price_color')
````

````python
soup.select("div p.price_color")
````

````python
print(soup.find_all('p', class_='price_color')[0].text)
sample_price = soup.find_all('p', class_='price_color')[0].text
float(sample_price[1:])
````

````python
def retrieve_prices(soup):
    trace = soup.find_all('p', class_ = 'price_color')
    prices = [float(p.text[1:]) for p in trace]
    return prices
````

````python
retrieve_prices(soup)
````

## Retrieve Availability

Write a function to retrieve whether each book is available or not. The function should take in the `soup` from a given html page and return a list of the availability for each book.

````python
soup.find_all('p', class_='instock availability')[0]
````

````python
print(soup.find_all('p', class_='instock availability')[0].text)
soup.find_all('p', class_='instock availability')[0].text.strip()
# clean all spaces/ non-necessary characters
````

````python
def retrieve_availabilities(soup):
    availabilities = [a.text.strip() for a in soup.find_all('p', class_='instock availability')]
    return availabilities
````

````python
retrieve_availabilities(soup)
````

## Create a Script to Retrieve All the Books From All 50 Pages

Finally, write a script to retrieve all of the information from all 50 pages of the books.toscrape.com website. 

````python
import pandas as pd
df = pd.DataFrame({"title": retrieve_titles(soup), "rating": retrieve_ratings(soup), "price": retrieve_prices(soup), "availability": retrieve_availabilities(soup)})
````

````python
for i in range(2,51):
    url = "http://books.toscrape.com/catalogue/page-{}.html".format(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    new_titles = retrieve_titles(soup)
    new_star_ratings = retrieve_ratings(soup)
    new_prices = retrieve_prices(soup)
    new_avails = retrieve_availabilities(soup)
    df2 = pd.DataFrame({"title": new_titles, "rating": new_star_ratings, "price": new_prices, "availability": new_avails})
    df = df.append(df2, ignore_index=True)
    print(df.shape)
````
